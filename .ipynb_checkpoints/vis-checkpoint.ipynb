```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
a_80_b_20_algo_metropolis_beta_1_n0_1 = pd.read_pickle('./data/a_80_b_20_algo_metropolis_beta_1_n0_1.pickle')
a_80_b_20_algo_houdayer_beta_1_n0_1 = pd.read_pickle('./data/a_80_b_20_algo_houdayer_beta_1_n0_1.pickle')
a_80_b_20_algo_houdayer_beta_1_n0_2 = pd.read_pickle('./data/a_80_b_20_algo_houdayer_beta_1_n0_2.pickle')
a_80_b_20_algo_houdayer_beta_1_n0_20 = pd.read_pickle('./data/a_80_b_20_algo_houdayer_beta_1_n0_20.pickle')
```

```python
m1 = list(filter(lambda c: type(c) == int, a_80_b_20_algo_metropolis_beta_1_n0_1['first_perfect_1']))
h1_1 = list(filter(lambda c: type(c) == int,a_80_b_20_algo_houdayer_beta_1_n0_1['first_perfect_1']))
h1_2 = list(filter(lambda c: type(c) == int,a_80_b_20_algo_houdayer_beta_1_n0_1['first_perfect_2']))
h2_1 = list(filter(lambda c: type(c) == int,a_80_b_20_algo_houdayer_beta_1_n0_2['first_perfect_1']))
h2_2 = list(filter(lambda c: type(c) == int,a_80_b_20_algo_houdayer_beta_1_n0_2['first_perfect_2']))
h3_1 = list(filter(lambda c: type(c) == int,a_80_b_20_algo_houdayer_beta_1_n0_20['first_perfect_1']))
h3_2 = list(filter(lambda c: type(c) == int,a_80_b_20_algo_houdayer_beta_1_n0_20['first_perfect_2']))
```

```python
D = [m1, h2_1, h2_2, h3_1, h3_2]
fig, ax = plt.subplots()

vp = ax.violinplot(D,[2, 4, 6, 8, 10], widths=2,
                   showmeans=True, showmedians=False, showextrema=False)

fig.suptitle('')
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.stats.api as sms
import scipy.stats as st
```

```python

d={}
for i in range(2,30): 
    d[f'n0_{i}']  = list(filter(lambda c: type(c) == \
        int,pd.read_pickle( f'./data/a_20_b_5_algo_houdayer_beta0_1_betaN_None_n0_{i+1}.pickle')['first_perfect_1']))

m1 = list(filter(lambda c: type(c) == int, pd.read_pickle( './data/a_20_b_5_algo_metropolis_beta0_1_betaN_None_n0_1.pickle')['first_perfect_1']))

D = []
for key, value in d.items():
    if len(value)> 0:
        D.append(value)
```

```python

fig, ax = plt.subplots(figsize=(20, 10))

vp = ax.violinplot(D, range(2,len(D)+2,1), widths=1,
                   showmeans=True, showmedians=False, showextrema=False)

met = np.mean(m1)
ax.axline((-10, met), (40, met), linewidth=0.5, color='r')
ax.set_xlim(-2, 32)
ax.violinplot(m1, [0], widths=1, showmeans=True,showmedians=False, showextrema=False)

ax.set_xlim(-1, 30)
ax.set_ylim(0,2000)


ax.set_xticks(range(2,31))

ax.set_xlabel('n_0')
ax.set_ylabel('Counts')
```

```python
ov={}
for i in [2,4,9]: 
    ov[f'n0_{i}']  = np.array(pd.read_pickle( f'./data/a_20_b_5_algo_houdayer_beta0_1_betaN_None_n0_{i}.pickle')['estimation_overlap_over_runs_1'])
for i in [50,100]: 
     ov[f'n0_{i}']  = np.array(pd.read_pickle( f'./data/a_20_b_5_N_100_it_5000_algo_houdayer_beta0_1_betaN_None_n0_{i}.pickle')['estimation_overlap_over_runs_1'])

m1 = np.array( pd.read_pickle( './data/a_20_b_5_algo_metropolis_beta0_1_betaN_None_n0_1.pickle')['estimation_overlap_over_runs_1'])
```

```python
# plot
fig, ax = plt.subplots(figsize=(15,10))
x = range(1,5001)

#ax.fill_between(x, y1, y2, alpha=.5, linewidth=0)
ax.plot(x, np.mean(m1, axis = 0), linewidth=1, label='Metropolis')
ci = st.t.interval(0.95, len(m1)-1, loc=np.mean(m1, axis = 0), scale=st.sem(m1))
ax.fill_between(x, ci[0], ci[1], alpha=.3, linewidth=0)

ax.plot(x, np.mean(ov[f'n0_2'], axis = 0), linewidth=1, label='Houdayer n0 = 2')
ci = st.t.interval(0.95, len(ov[f'n0_2'])-1, loc=np.mean(ov[f'n0_2'], axis = 0), scale=st.sem(ov[f'n0_2']))
ax.fill_between(x, ci[0], ci[1], alpha=.3, linewidth=0)

ax.plot(x, np.mean(ov[f'n0_4'], axis = 0), linewidth=1, label='Mixed Houdayer n0 = 4')
ci = st.t.interval(0.95, len(ov[f'n0_4'])-1, loc=np.mean(ov[f'n0_4'], axis = 0), scale=st.sem(ov[f'n0_4']))
ax.fill_between(x, ci[0], ci[1], alpha=.3, linewidth=0)

ax.plot(x, np.mean(ov[f'n0_9'], axis = 0), linewidth=1, label='Mixed Houdayer n0 = 9')
ci = st.t.interval(0.95, len(ov[f'n0_9'])-1, loc=np.mean(ov[f'n0_9'], axis = 0), scale=st.sem(ov[f'n0_9']))
ax.fill_between(x, ci[0], ci[1], alpha=.3, linewidth=0)

ax.plot(x, np.mean(ov[f'n0_50'], axis = 0), linewidth=1, label='Mixed Houdayer n0 = 50')
ci = st.t.interval(0.95, len(ov[f'n0_50'])-1, loc=np.mean(ov[f'n0_50'], axis = 0), scale=st.sem(ov[f'n0_50']))
ax.fill_between(x, ci[0], ci[1], alpha=.3, linewidth=0)

#ax.plot(x, np.mean(ov[f'n0_100'], axis = 0), linewidth=1, label='Mixed Houdayer n0 = 100')
#ci = st.t.interval(0.95, len(ov[f'n0_100'])-1, loc=np.mean(ov[f'n0_100'], axis = 0), scale=st.sem(ov[f'n0_100']))
#ax.fill_between(x, ci[0], ci[1], alpha=.3, linewidth=0)

ax.set_xlim(0, 2000)
ax.set_ylim(0,1.05)
ax.legend()

ax.set_xlabel('Number of steps')
ax.set_ylabel('Overlap')
```

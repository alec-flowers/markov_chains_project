```python
import numpy as np
def calculate_beta(f, l, t, T):
    t_f = 1/f
    t_l = 1/l
    t0 = t_f*t_l*(np.log(T+1) - np.log(2))
    t1 = (t_l*np.log(T+1)-t_f*np.log(2))+(t_f-t_l)*np.log(t+1)
    beta = t0/t1
    return 1/beta
```

```python
f = .4
l = 1
T = 1000
schedule = []
for t in range(T):
    schedule.append(calculate_beta(f, l, t+1, T))
```

```python
plt.plot(schedule)
```
